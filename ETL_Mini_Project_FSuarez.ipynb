### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
crowdfunding_info_df.columns
```

```python
# Assign the category and subcategory values to category and subcategory columns.
# Split the 'category & sub-category' column into 'category' and 'subcategory' columns
crowdfunding_info_df[['category', 'subcategory']] = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)

# Display the updated DataFrame with the new columns
crowdfunding_info_df.head()
```

```python
# Get the unique categories and subcategories in separate lists.
categories = crowdfunding_info_df['category'].unique().tolist()
subcategories = crowdfunding_info_df['subcategory'].unique().tolist()
print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 
cat_ids = ['cat' + str(cat) for cat in category_ids]

# Use a list comprehension to add "subcat" to each subcategory_id.    
scat_ids = [f"subcat{subcat_id}" for subcat_id in subcategory_ids]

print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame({'category_id': category_ids, 'category': categories})
# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_df = pd.DataFrame({'subcategory_id': subcategory_ids, 'subcategory': subcategories})
```

```python
# Set the category_id as the index
#Reset the Index
category_df.reset_index(inplace=True, drop=True)

#adding the cat prefix to the category_id
category_df['category_id'] = ['cat' + str(i+1) for i in range(len(category_df))]

# Display the DataFrame with the new category_id column
category_df
```

```python
#Fixing the numbers from subcategory_id to right position
subcategory_df.reset_index(inplace=True, drop=True)

#adding the subcat prefix to the subcategory_id
subcategory_df['subcategory_id'] = ['subcat' + str(i+1) for i in range(len(subcategory_df))]

# Display the DataFrame with the new subcategory_id column
subcategory_df
```

```python
# Export categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)

subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

**Create a Campaign DataFrame that has the following columns:**

```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
campaign_df.rename(columns={'blurb': 'campaign', 'launched_at': 'launch_date', 'deadline': 'end_date'}, inplace=True)
```

```python
# Merge campaign_df with category_df on the 'category' column
campaign_df = campaign_df.merge(category_df, how='left', left_on='category', right_on='category')

# Merge the result with subcategory_df on the 'subcategory' column
campaign_df = campaign_df.merge(subcategory_df, how='left', left_on='subcategory', right_on='subcategory')

# Display the updated DataFrame to confirm merges
campaign_df.head()
```

```python
# Check the datatypes
dtype_df = campaign_df.dtypes
print(dtype_df)
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt

# Convert the 'launch_date' and 'end_date' columns to datetime format
campaign_df['launch_date'] = pd.to_datetime(campaign_df['launch_date'], unit='s')
campaign_df['end_date'] = pd.to_datetime(campaign_df['end_date'], unit='s')

```

```python
campaign_df = campaign_df.merge(category_df, how='left', left_on='category_id', right_on='category_id')
campaign_df = campaign_df.merge(subcategory_df, how='left', left_on='subcategory_id', right_on='subcategory_id')
campaign_df.head()

```

```python
#CHECKING THE DATAFRAME AFTER MERGING THE DATAFRAMES

#Checking the columns 
print(campaign_df.columns)
#Checking the datatypes of the columns
print(campaign_df.dtypes)
#Checking the number of columns
print(len(campaign_df.columns))
```

```python
# Drop unwanted columns
columns_to_drop = ['staff_pick', 'spotlight', 'category & sub-category']
campaign_df = campaign_df.drop(columns=columns_to_drop)

# Display the DataFrame after dropping unwanted columns
campaign_df.head()
```

```python
# Drop unwanted columns

# Identify columns to drop
columns_to_drop = ['category_x', 'subcategory_x', 'category_y', 'subcategory_y']
campaign_df = campaign_df.drop(columns=columns_to_drop)

# Rename columns as required
campaign_df.rename(columns={'campaign': 'description'}, inplace=True)

# Display the cleaned DataFrame
campaign_df.head()
```
